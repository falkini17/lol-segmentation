# fetch_game_stats()

```python
def fetch_game_stats(region, game_id):
    participants = []
    region = region
    game_id = game_id
    game_stats = watcher.match.by_id(region, game_id)
    game_duration = game_stats['info']['gameDuration']/1000
    game_mode = game_stats['info']['gameMode']
    game_type = game_stats['info']['gameType']
    for row in game_stats['info']['participants']:
        participants_row = {}
#         participants_row['participantId'] = row['participantId']
        participants_row['lane'] = row['lane']
        participants_row['role'] = row['role']
        participants_row['teamId'] = row['teamId']
        participants_row['champion'] = row['championId']
        participants_row['summonerLevel'] = row['summonerLevel']
#         participants_row['spell1'] = row['spell1Id']
#         participants_row['spell2'] = row['spell2Id']
        participants_row['win'] = row['win']
        participants_row['kills'] = row['kills']
        participants_row['deaths'] = row['deaths']
        participants_row['assists'] = row['assists']
        participants_row['totalDamageDealt'] = row['totalDamageDealt']
        participants_row['totalDamageDealtToChampions'] = row['totalDamageDealtToChampions']
        participants_row['totalDamageTaken'] = row['totalDamageTaken']
        participants_row['damageSelfMitigated'] = row['damageSelfMitigated']
        participants_row['magicDamageDealt'] = row['magicDamageDealt']
        participants_row['physicalDamageDealt'] = row['physicalDamageDealt']
        participants_row['goldEarned'] = row['goldEarned']
        participants_row['goldSpent'] = row['goldSpent']
        participants_row['inhibitorKills'] = row['inhibitorKills']
        participants_row['champLevel'] = row['champLevel']
        participants_row['totalMinionsKilled'] = row['totalMinionsKilled']
        try:
            participants_row['wardsPlaced'] = row['wardsPlaced']
        except KeyError:
            participants_row['wardsPlaced'] = 0
        participants_row['longestTimeSpentLiving'] = row['longestTimeSpentLiving']
        participants_row['totalTimeCrowdControlDealt'] = row['totalTimeCCDealt']
        participants_row['item0'] = row['item0']
        participants_row['item1'] = row['item1']
        participants_row['item2'] = row['item2']
        participants_row['item3'] = row['item3']
        participants_row['item4'] = row['item4']
        participants_row['item5'] = row['item5']
        participants_row['item6'] = row['item6']
        participants_row['champExperience'] = row['champExperience'] #dodane z match_api_v5
        participants_row['damageDealtToTurrets'] = row['damageDealtToTurrets'] #dodane z match_api_v5
        participants_row['gameEndedInSurrender'] = row['gameEndedInSurrender'] #dodane z match_api_v5
        participants_row['largestMultiKill'] = row['largestMultiKill'] #dodane z match_api_v5
        participants_row['spell4Casts'] = row['spell4Casts'] #dodane z match_api_v5
        participants_row['totalHeal'] = row['totalHeal'] #dodane z match_api_v5
        participants_row['wardsKilled'] = row['wardsKilled'] #dodane z match_api_v5
        participants_row['individualPosition'] = row['individualPosition'] #dodane z match_api_v5
        participants_row['teamPosition'] = row['teamPosition'] #dodane z match_api_v5
        participants_row['championName'] = row['championName'] #dodane z match_api_v5
        participants_row['turretKills'] = row['turretKills'] #dodane z match_api_v5
        participants.append(participants_row)
    df = pd.DataFrame(participants)
    df['gameId'] = game_id
    df['gameDurationSec'] = game_duration
    df['gameDurationMin'] = game_duration/60
    df['gameMode'] = game_mode
    df['gameType'] = game_type
    df = df[df['gameMode']=='CLASSIC']
    df['playerRank'] = df.sort_values(['champLevel','champExperience'], ascending=False)\
             .groupby(['teamId'])\
             .cumcount() + 1
    df['killsPerMinute'] = df['kills']/df['gameDurationMin']
    df['deathsPerMinute'] = df['deaths']/df['gameDurationMin']
    df['assistsPerMinute'] = df['assists']/df['gameDurationMin']
    df['damageDealtPerMinute'] = df['totalDamageDealt']/df['gameDurationMin']
    df['damageDealtToChampionsPerMinute'] = df['totalDamageDealtToChampions']/df['gameDurationMin']
    df['damageTakenPerMinute'] = df['totalDamageTaken']/df['gameDurationMin']
    df['damageSelfMitigatedPerMinute'] = df['damageSelfMitigated']/df['gameDurationMin']
    df['magicDamageDealtPerMinute'] = df['magicDamageDealt']/df['gameDurationMin']
    df['physicalDamageDealtPerMinute'] = df['physicalDamageDealt']/df['gameDurationMin']
    df['goldEarnedPerMinute'] = df['goldEarned']/df['gameDurationMin']
    df['goldSpentPerMinute'] = df['goldSpent']/df['gameDurationMin']
    df['minionsKilledPerMinute'] = df['totalMinionsKilled']/df['gameDurationMin']
    df['wardsPlacedPerMinute'] = df['wardsPlaced']/df['gameDurationMin']
    df['wardsKilledPerMinute'] = df['wardsKilled']/df['gameDurationMin']
    df['inhibitorKillsPerMinute'] = df['inhibitorKills']/df['gameDurationMin']
    df['timeCrowdControlDealtPerMinute'] = df['totalTimeCrowdControlDealt']/df['gameDurationMin']
    df['kdRatio'] = df['kills']/(df['deaths']+1)
    df['percentDamageDealtToChampions'] = df['totalDamageDealtToChampions']/df['totalDamageDealt']*100
    df['damageRatio'] = df['totalDamageDealt']/df['totalDamageTaken']
    df['goldRatio'] = df['goldSpent']/df['goldEarned']
    df['percentMagic'] = df['magicDamageDealt']/df['totalDamageDealt']*100
    df['percentPhysical'] = df['physicalDamageDealt']/df['totalDamageDealt']*100
    df['technicalItems'] = df['item0'].astype('str') + '  ' + df['item1'].astype('str') + '  ' + df['item2'].astype('str') + '  ' + df['item3'].astype('str')\
    + '  ' + df['item4'].astype('str') + '  ' + df['item5'].astype('str') + '  ' + df['item6'].astype('str')
    df['emptyItemSlots'] = df['technicalItems'].str.count(' 0 ')
    df['role'].replace({'NONE':'JUNGLE'}, inplace=True)
    df['longestTimeSpentLiving'] = df['longestTimeSpentLiving']/60 
    df['longestTimeSpentLiving'] = np.where(df['longestTimeSpentLiving']==0,df['gameDurationMin'],df['longestTimeSpentLiving'])
    df['experiencePerMinute'] = df['champExperience']/df['gameDurationMin']
    df['damageDealtToTurretsPerMinute'] = df['damageDealtToTurrets']/df['gameDurationMin']
    df['ultimateCastsPerMinute'] = df['spell4Casts']/df['gameDurationMin']
    df['totalHealPerMinute'] = df['totalHeal']/df['gameDurationMin']
    df['turretKillsPerMinute'] = df['turretKills']/df['gameDurationMin']
    df = df[[
        'gameId'
        ,'teamId'
        ,'lane'
        ,'role'
        ,'individualPosition'
        ,'teamPosition'
        ,'summonerLevel'
        ,'playerRank'
        ,'gameDurationMin'
        ,'win'
        ,'gameEndedInSurrender'
        ,'champLevel'
        ,'championName'
        ,'experiencePerMinute'
        ,'killsPerMinute'
        ,'deathsPerMinute'
        ,'assistsPerMinute'
        ,'damageDealtPerMinute'
        ,'damageDealtToChampionsPerMinute'
        ,'damageTakenPerMinute'
        ,'damageSelfMitigatedPerMinute'
        ,'magicDamageDealtPerMinute'
        ,'physicalDamageDealtPerMinute'
        ,'damageDealtToTurretsPerMinute'
        ,'totalHealPerMinute'
        ,'goldEarnedPerMinute'
        ,'goldSpentPerMinute'
        ,'minionsKilledPerMinute'
        ,'wardsPlacedPerMinute'
        ,'wardsKilledPerMinute'
        ,'inhibitorKillsPerMinute'
        ,'turretKillsPerMinute'
        ,'timeCrowdControlDealtPerMinute'
        ,'ultimateCastsPerMinute'
        ,'kdRatio'
        ,'percentDamageDealtToChampions'
        ,'damageRatio'
        ,'goldRatio'
        ,'percentMagic'
        ,'percentPhysical'
        ,'longestTimeSpentLiving'
        ,'emptyItemSlots'
        ,'largestMultiKill'
    ]]
    return df
```

# fetch_game_history()

```python
def fetch_game_history(region_old,region_new, login):
    region_old = region_old
    region_new = region_new
    playerName = login
    identity = watcher.summoner.by_name(region_old, playerName)
    puuid = identity['puuid']
    matches = watcher.match.matchlist_by_puuid(region_new, puuid)
    matches = pd.DataFrame(matches,columns=['gameId'])
    matches['region']=region_new
    return matches
```

# fetch_game_participants_from_game_id()

```python
def fetch_game_participants_from_game_id(game_id, region):
    region = region
    game_id = game_id
    game_stats = watcher.match.by_id(region, game_id)
    participants = []
    for row in game_stats['info']['participants']:
        participants_row = {}
        participants_row['puuid'] = row['puuid']
        participants_row['summonerName'] = row['summonerName']
        participants_row['gameId'] = game_id
        participants_row['region'] = region
        participants.append(participants_row)
        df = pd.DataFrame(participants)
    return df
```

```python
def get_all_stats_by_player_name(old_region,new_region,login):
    
    start_time = datetime.now() 
    rate_limiter = RateLimiter(max_calls=19, period=1)
    game_history = fetch_game_history(old_region,new_region,login)
    appended_df = []
    for i in game_history.index:
        with rate_limiter:   
            dataframe = fetch_game_stats(game_history.iat[i,1],game_history.iat[i,0])
            appended_df.append(dataframe)
    appended_df = pd.concat(appended_df)
    print('Time elapsed (hh:mm:ss.ms) {}'.format(datetime.now() - start_time))
    return appended_df
```

```python
def fetch_all_ever_game_participants_by_player_name(region_old,region_new,login):
    
    start_time = datetime.now() 
    rate_limiter = RateLimiter(max_calls=19, period=1)
    player_history = fetch_game_history(region_old,region_new,login)
    appended_df = []
    for i in player_history.index:
        dataframe = fetch_game_participants_from_game_id(player_history.iat[i,0],player_history.iat[i,1])
        appended_df.append(dataframe)
    x = pd.concat(appended_df)
    x['old_region'] = x['gameId'].str[:4].str.lower()
    x = x[['summonerName','old_region','region']]
    x = x.drop_duplicates().reset_index(drop=True)
    print('Time elapsed (hh:mm:ss.ms) {}'.format(datetime.now() - start_time))
    return x
```

```python
def get_final_dataframe(region_old,region_new,login):
    start_time = datetime.now() 
    rate_limiter = RateLimiter(max_calls=19, period=1)
    player_participants = fetch_all_ever_game_participants_by_player_name(region_old,region_new,login)
    appended_df = []
    for i in player_participants.index:
        try:
            print(i)
            dataframe = get_all_stats_by_player_name(player_participants.iat[i,1],player_participants.iat[i,2],player_participants.iat[i,0])
            appended_df.append(dataframe)
        except:
            pass
    x = pd.concat(appended_df)
#     x = x.drop_duplicates()
    print('Time elapsed (hh:mm:ss.ms) {}'.format(datetime.now() - start_time))
    return x
```
